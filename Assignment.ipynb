```python
import numpy as np
import pandas as pd
from random import randrange
```

# a) part

```python
num_of_labels = int(input('number of labels: '))
distribution = [int(x) for x in (input('distribution : ')).split()]
```

```python
# Generating dataset
def generate_dataset(num_of_labels, distribution, length=10000):
    # creating dataset from uniform distribution 
    dataset = pd.DataFrame(np.random.uniform(low=-1, high=1, size=length), columns=['Feature'], dtype=np.float64)
    
    # Generating class labels and stacking it with feature
    x = []
    for i,dist in enumerate(distribution):
        x.extend([i+1]*(dist*100))
    dataset['Target'] = pd.DataFrame(np.asarray(x), dtype=np.int64)

    return dataset
```

```python
dataset = generate_dataset(num_of_labels, distribution)
dataset.head()
```

```python
dataset['Target'].value_counts()
```

# b) part

```python
# Split a dataset into k=10 folds
def cross_validation_split(dataset, num_of_labels, distribution, folds=10):
    
    sum = 0
    index = []
    for d in distribution:
        index.append(sum*100)
        sum+=d
        
    dataset_split = list()
    dataset_copy = list(dataset.values.tolist())
    fold_size = int(dataset.shape[0] / folds)  # length / k-folds
    
    # Creating stratified 10 folds 
    for i in range(folds):
        fold = list()
        for j, dist in enumerate(distribution):
            fold.extend(dataset_copy[index[j]+i*dist*10 : index[j]+i*dist*10+dist*10])
        dataset_split.append(fold)
    return dataset_split
```

```python
folds = cross_validation_split(dataset, num_of_labels, distribution) 
for i, fold in enumerate(folds):
    print('\n *************** ', i+1,' Fold ****************************************' )
    for f in fold:
        print(f)
    print('\n **********************************************************************' )
```
